```python
"""Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions."""
```

```python
import seaborn as sns
```

```python
fmri= sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot(x =fmri.timepoint , y =fmri.signal)
```

```python
"""Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset."""
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
x= flights.pivot_table(index= 'year', columns= 'month', values= 'passengers' ,aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False, linewidths= 0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths= 0.5, annot= True, fmt= 'd')
```
